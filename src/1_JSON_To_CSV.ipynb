# Read & Store data

```python
import pandas as pd
import gzip
```

```python
# use python generator to iterate through books2 and reviews2
def parse(path):
    g = gzip.open(path, 'r')
    for l in g:
        yield eval(l)
```

```python
books2 = (parse('input/metadata/goodreads_books_children.json.gz'))
type(books2)
```

```python
bk_lst = list(books2)
len(bk_lst)
```

```python
bk_df = pd.DataFrame(bk_lst)
bk_df.to_csv (r'processed_data/books.csv', index = None)
```

```python
bk_df.head(2)
```

```python
reviews = (parse('input/metadata/goodreads_reviews_children.json.gz'))
next(reviews)
```

```python
rev_lst = list(reviews)
len(rev_lst)
```

```python
reviews_df = pd.DataFrame(rev_lst)
reviews_df.to_csv (r'processed_data/reviews.csv', index = None)
```
